Au lieu d'Ã©crire :

### ğŸ“– B. Gestion d'erreurs pythonique : EAFP

**EAFP** = "Easier to Ask Forgiveness than Permission"

En Python, on prÃ©fÃ¨re **essayer puis gÃ©rer l'erreur** plutÃ´t que **vÃ©rifier avant**.

````python
donnees = {"nom": "Alice", "age": 25}

# âŒ Style "Look Before You Leap" (Java/C)
if "email" in donnees:
    email = donnees["email"]
else:
    email = "inconnu@example.com"
print(f"Style LBYL : {email}")

# âœ… Style Python : EAFP
try:
    email = donnees["email"]
except KeyError:
    email = "inconnu@example.com"
print(f"Style EAFP : {email}")

# ğŸ¯ Encore mieux : dict.get()
email = donnees.get("email", "inconnu@example.com")
print(f"Style pythonique : {email}")
````

````python
def est_palindrome(texte):
````

```python
    return normalise == normalise[::-1]
```

```python
def compresser_rle(texte):
    """
    Compresse un texte avec l'algorithme RLE.
    
    Args:
        texte: ChaÃ®ne Ã  compresser
        
    Returns:
        str: ChaÃ®ne compressÃ©e (format: caractÃ¨re + nombre de rÃ©pÃ©titions)
    """
    if not texte:
        return ""
    
    resultat = []
    i = 0
    
    while i < len(texte):
        caractere = texte[i]
        compte = 1
        
        # Compter les rÃ©pÃ©titions consÃ©cutives
        while i + compte < len(texte) and texte[i + compte] == caractere:
            compte += 1
        
        resultat.append(f"{caractere}{compte}")
        i += compte
    
    return ''.join(resultat)
```
